## 8958번 - OX퀴즈

### 문제

  "OOXXOXXOOO"와 같은 OX퀴즈의 결과가 있다. O는 문제를 맞은 것이고, X는 문제를 틀린 것이다. 문제를 맞은 경우 그 문제의 점수는 그 문제까지 연속된 O의 개수가 된다. 예를 들어, 10번 문제의 점수는 3이 된다.

  "OOXXOXXOOO"의 점수는 1+2+0+0+1+0+0+1+2+3 = 10점이다.

  OX퀴즈의 결과가 주어졌을 때, 점수를 구하는 프로그램을 작성하시오.

- 입력

  첫째 줄에 테스트 케이스의 개수가 주어진다. 각 테스트 케이스는 한 줄로 이루어져 있고, 길이가 0보다 크고 80보다 작은 문자열이 주어진다. 문자열은 O와 X만으로 이루어져 있다.

- 출력

  각 테스트 케이스마다 점수를 출력한다.

### 입출력 예시

- 예제 입력 1
```
5
OOXXOXXOOO
OOXXOOXXOO
OXOXOXOXOXOXOX
OOOOOOOOOO
OOOOXOOOOXOOOOX
```
- 예제 출력 1
```
10
9
7
55
30
```

### 풀이

1. O를 몇번 연속으로 맞췄는지 계산한다
2. 맞췄다면 가산하고, 틀렸다면 연속 정답 보너스를 초기화한다

````python
n = int(input()) # input line 수

for _ in range(n):
  line = input() # 문자열 입력받기

  acc = 1 # 연속 정답 보너스 계산용
  score = 0 # 총점

  for a in line: # 문자열에서 정오답 결과를 하나씩 확인
    if a == "O": # 정답이면
      score += acc # 점수에 보너스 더하기
      acc += 1 # 보너스 더하기
    else:
      acc = 1 #보너스 초기화
  print(score) # 정답 출력
````

## 1259번 - 팰린드롬수

### 문제

  어떤 단어를 뒤에서부터 읽어도 똑같다면 그 단어를 팰린드롬이라고 한다. 'radar', 'sees'는 팰린드롬이다.

  수도 팰린드롬으로 취급할 수 있다. 수의 숫자들을 뒤에서부터 읽어도 같다면 그 수는 팰린드롬수다. 121, 12421 등은 팰린드롬수다. 123, 1231은 뒤에서부터 읽으면 다르므로 팰린드롬수가 아니다. 또한 10도 팰린드롬수가 아닌데, 앞에 무의미한 0이 올 수 있다면 010이 되어 팰린드롬수로 취급할 수도 있지만, 특별히 이번 문제에서는 무의미한 0이 앞에 올 수 없다고 하자.

- 입력

  입력은 여러 개의 테스트 케이스로 이루어져 있으며, 각 줄마다 1 이상 99999 이하의 정수가 주어진다. 입력의 마지막 줄에는 0이 주어지며, 이 줄은 문제에 포함되지 않는다.

- 출력

  각 줄마다 주어진 수가 팰린드롬수면 'yes', 아니면 'no'를 출력한다.

### 입출력 예시

- 예제 입력 1
```
121
1231
12421
0
```
- 예제 출력 1
```
yes
no
yes
```

### 풀이

1. 파이썬에서 제공하는 기능으로 문자열을 뒤집어 원본과 같은지 확인한다

````python
# 풀이 1
while True:
  a = input() # 입력 숫자(문자열)

  if a == '0': # 종료 조건
    break

  print('yes' if a == a[::-1] else 'no') # 삼항연산: a reverse == a이면 yes, 아니면 no
````

1. 입력받은 수(str)의 양쪽이 같은지 확인한다.
  1. 같다면 양쪽을 잘라 길이를 줄이고 1.로 돌아간다.
  2. 다르다면 팰린드롬수가 아니다
2. 1.을 문자열이 남아있을 때까지 반복한다. 문자열이 빈 문자열이 되었다면 팰린드롬수이다.


````python
# 풀이 2
while True:
  a = input()
  if a == '0':
    break

  while a:
    if a[0] == a[-1]: # 양쪽 끝이 같은지 확인
      a = a[1:-1] # 같으면 양쪽 끝을 잘라냄
    else:
      print('no') # 다르면 종료
      break
  else:
    print('yes') # break 없이 종료된 while문, 따라서 팰린드롬수임
````

## 1978번 - 소수 찾기

### 문제

  주어진 수 N개 중에서 소수가 몇 개인지 찾아서 출력하는 프로그램을 작성하시오.

- 입력

  첫 줄에 수의 개수 N이 주어진다. N은 100이하이다. 다음으로 N개의 수가 주어지는데 수는 1,000 이하의 자연수이다.

- 출력

  주어진 수들 중 소수의 개수를 출력한다.

### 입출력 예시

- 예제 입력 1
```
4
1 3 5 7
```
- 예제 출력 1
```
3
```

### 풀이

1. 받은 소수 후보 목록을 리스트로 저장해, for문으로 반복한다
2. 소수인지 확인한다
  1. 1은 소수가 아니므로 제외한다
  2. [2, √n) 범위의 정수로 나뉘어떨어지면 소수가 아니다.
  3. 나뉘어떨어지지 않았으면 소수이다.
3. 소수의 총 개수를 출력한다.

````python
_ = int(input()) # 숫자의 개수(파이썬에서 사용안함)
numlist = map(int, input().split()) # 입력받은 숫자들을 리스트로 저장
result = 0 # 소수의 수

for i in numlist:

  if i == 1: # 1은 소수가 아님; for문에 진입 불가해 소수로 처리됨
    continue

  for j in range(2, int(i**(1/2) + 1)):
    if i % j == 0:
      break # 소수 검사: [2, i^(1/2)] 범위에서 약수가 존재하면 소수가 아님
  else:
    result += 1 # 나뉘지 않았으면 소수임

print(result) # 결과 출력
````
